```python
# Artificial Intelligence Project 3
# Loading the necessary Pakages
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# PHASE 1: THE DATASET (KNOWLEDGE BASE)
# This acts as our database of items to recommend.
# In a real company, this would be thousands of rows from a database.
data = {
    "Career_Path": [
        "Data Scientist",
        "Frontend Developer",
        "Cybersecurity Analyst",
        "Backend Engineer",
        "UI/UX Designer",
        "Cloud Architect"
    ],
    "Keywords": [
        "math statistics artificial intelligence python machine learning predict models data data analysis",
        "design interfaces css html javascript user experience visual responsive websites react frontend",
        "networks security hackers protect firewall encryption vulnerabilities cyber defend defense",
        "servers database logic api optimization sql java architecture infrastructure backend scalable",
        "wireframes prototyping empathy user research figma visual design usability layout interface",
        "aws azure scalability deployment servers hosting architecture distributed systems cloud"
    ],
    "Suggested_Project": [
        "Build a machine learning model to predict housing prices.",
        "Code a responsive personal portfolio website.",
        "Create a script to audit network vulnerabilities.",
        "Build a RESTful API for a to-do list application.",
        "Design a high-fidelity prototype for a delivery app in Figma.",
        "Deploy a scalable web app using AWS EC2 and S3."
    ]
}

df = pd.DataFrame(data)


# PHASE 2: THE AI MATCHMAKING ENGINE

def matchmaker_engine(user_input, top_n=3):
    """
    Takes natural language user input, converts it to mathematical vectors,
    and uses Cosine Similarity to find the best career matches.
    """
    print(f"\nAnalyzing User Intent: '{user_input}'\n")
    print("-" * 50)

    # 1. Feature Extraction (Text to Vectors)
    # TF-IDF turns words into numbers, giving more weight to unique/important words
    vectorizer = TfidfVectorizer(stop_words='english')

    # We combine the user's input with our database to map them in the same mathematical space
    all_text_data = [user_input] + df['Keywords'].tolist()
    vector_matrix = vectorizer.fit_transform(all_text_data)

    # 2. The Math Engine (Cosine Similarity)
    # We measure the angle between the user's vector (index 0) and all other career vectors
    similarity_scores = cosine_similarity(vector_matrix[0:1], vector_matrix[1:]).flatten()

    # Add the scores to our dataframe so we can sort them
    df['Similarity_Score'] = similarity_scores

    # 3. AI Prediction Logic (Sorting the best matches)
    top_matches = df.sort_values(by='Similarity_Score', ascending=False).head(top_n)

    # 4. Output Generation
    for index, row in top_matches.iterrows():
        # Convert decimal to percentage for readability
        match_percentage = round(row['Similarity_Score'] * 100, 1)

        # Only show matches that actually have some relevance (score > 0)
        if match_percentage > 0:
            # Create a cool visual progress bar for the terminal
            bar = "█" * int(match_percentage / 5)

            print(f"🎯 Match: {row['Career_Path']} ({match_percentage}%)")
            print(f"📊 Confidence: |{bar.ljust(20, '-')}|")
            print(f"🚀 Recommended Project: {row['Suggested_Project']}")
            print("-" * 50)


# PHASE 3: RAW DATA INPUT (TESTING)

# Let's test it with a few different student profiles!

# Test Profile 1: The Math/AI enthusiast
test_input_1 = "I really like math, finding patterns in numbers, and building artificial intelligence models."
matchmaker_engine(test_input_1, top_n=2)

# Test Profile 2: The Visual/Creative student
test_input_2 = "I love making things look beautiful. I care about user experience, colors, and building responsive websites."
matchmaker_engine(test_input_2, top_n=2)
```
